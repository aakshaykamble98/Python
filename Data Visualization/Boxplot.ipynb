# Boxplot
- A boxplot is a standerdized way of displaying the distribution of data based on five numbers summary ("minimum", First Quartile (Q1), Median, Third Quartile (Q3), "maximum"). It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed. 



**Comparison between Boxplot and Normal Distribution plot**



```python
#Import necessary libraries

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('cars.csv')
df
```

```python
df.describe()
```

```python
df.describe(include='O')
```

```python
#Set the figure size of the graph

plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['figure.dpi']=200
```

```python
df.info()
```

```python
#Selecting numerical columns

num_cols = df.select_dtypes([float, 'int64'])
num_cols
```

```python
#Matplotlib Boxplot

plt.boxplot(df['symboling'], notch=True,)
```

```python
plt.boxplot(df['symboling'], notch=True, patch_artist=True)
```

```python
plt.boxplot(num_cols)
num_cols.head(2)
```

```python
num_cols.drop(columns=['price','engine-size']).boxplot()
```

```python
num_cols.drop(columns=['symboling','price','engine-size']).boxplot()
```

```python
plt.boxplot(num_cols['city-mpg'], notch=True, patch_artist=True)
plt.grid()
```
