# Subplots

- Subplots means group of axes that can exist in a single matplotlib figure. subplots() function in the matplotlib library, helps in creating multiple layouts of subplots. It provides control all over the individuel plots that are created.


- When analyzing data you might want to compare multiple plots placed side by side. Matplotlib provides a convenient method called subplots to do this.


- Subplots means a group of smaller axes (where each axes is a plot) that can exist together within a single figure. Think of figure as a canvas that holds multiple plots.

```python
plt.rcParams['figure.figsize']=(7,4)
plt.rcParams['figure.dpi']=250
```

```python
fig, ax =plt.subplots(2,2)
ax[0,0].plot(a,b)
ax[0,1].scatter(a,b)
ax[1,0].boxplot(a)
ax[1,1].plot(b,a)
```

```python
#Read external data

df= pd.read_csv('cars.csv')
```

```python
df.head(3)
```

```python
#Generate subplots
```

```python
plt.rcParams['figure.figsize']=(12,6)
plt.rcParams['figure.dpi']=300
```

```python
fig, axes= plt.subplots(2,2)

sns.distplot(df['width'], ax=axes[0,0])
sns.kdeplot(df['highway-mpg'],ax= axes[0,1])
sns.boxplot(df['price'],ax=axes[1,1])
sns.scatterplot(df['price'], df['engine-size'], hue=df['fuel-type'], ax=axes[1,0])

plt.tight_layout()
```
