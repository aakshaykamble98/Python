```python
#Import libraries

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings 
warnings.filterwarnings('ignore')
```

```python
#load the datasets

glaxo=pd.read_csv('GLAXO.csv')
glaxo
```

```python
beml=pd.read_csv('BEML.csv')
beml
```

```python
#Select relevent column

glaxo=glaxo[['Date','Close']]
beml=beml[['Date','Close']]
```

```python
glaxo
```

```python
beml
```

```python
#Data Discription
glaxo.describe()
```

```python
glaxo.info()
```

```python
beml.describe()
```

```python
beml.info()
```

```python
#Change datatype of date columns to datetime

glaxo['Date']=pd.DatetimeIndex(glaxo['Date'])
beml['Date']=pd.DatetimeIndex(beml['Date'])
```

```python
glaxo.info()
```

```python
beml.info()
```

```python
glaxo=glaxo.set_index(glaxo['Date'])
beml=beml.set_index(beml['Date'])
```

```python
glaxo
```

```python
beml
```

```python
glaxo.drop(columns='Date', inplace=True)
```

```python
beml.drop(columns='Date', inplace=True)
```

```python
glaxo
```

```python
beml
```

```python
#Plot closing value for both glaxo and beml

sns.set_style('darkgrid')
plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['figure.dpi']=250
plt.plot(glaxo['Close'])
plt.xlabel=('Time')
plt.ylabel=('Close')
```

```python
plt.plot(beml['Close'])
plt.xlabel=('Time')
plt.ylabel=('Close')
```

```python
#Calculate Gains

glaxo['gain']=glaxo['Close'].pct_change(periods=1)
beml['gain']=beml['Close'].pct_change(periods=1)
```

```python
glaxo
```

```python
beml
```

```python
#Drop null values
glaxo.dropna(inplace=True)
```

```python
beml.dropna(inplace=True)
```

```python
#Plot gain values

plt.plot(glaxo['gain'])
plt.axhline(glaxo['gain'].mean(),color='r')
plt.xlabel=('Time')
plt.ylabel=('Gain')
```

```python
plt.plot(beml['gain'])
plt.axhline(beml['gain'].mean(), color='y')
plt.xlabel=('Time')
plt.ylabel=('Gain')
```

```python
#subplot for distribution of beml and glaxo
```

```python
glaxo 
```

```python
plt.subplot(1,2,1)
sns.distplot(glaxo['gain'], color='r', label='GLAXO')

plt.legend()
plt.title('Distribution Graph')

plt.subplot(1,2,2)
sns.distplot(beml['gain'],color='b', label='BEML')

plt.legend()
plt.title('Distribution Graph')
```

```python
sns.distplot(glaxo['gain'], color='r', label='GLAXO')
plt.xlabel('Time')
plt.ylabel('Gain')
plt.legend()
plt.title('Distribution Graph')
```
