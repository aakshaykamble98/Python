#  Why to visualize data ?

- Data visualization is the prctice of translating information into a visual context, such as a map or a graph, to make data easier for the human brains to understand and pull insight from. The main gaol of the data visualization is to make it easier to identify patterns, trends and outliers in the large data sets.
    - Better Analysis
    - Quick Action
    - Identifying Patterns
    - Finding Errors
    - Understand the Stories
    - Exploring Bussiness Insights
    - Grasping the Latest Trend

# What is Matplotlib ?

- Matplotlib is a python visualization library, majorly used for plotting 2D data.
- Matplotlib is a multi platform data visualization library built on NumPy arrays and designed to work with the broder SciPy stack.
- Matplotlib consists of several plots like line, bar, scatter, histogram etc.

- Xlim: Used for define limit for x variable
- Ylim: Used for define limit for y variable


```python
area=[1000,1520,3000,4500,5200]
price=[20,25,35,42,55]
plt.plot(area,price,'darkred',linewidth=2.5)
plt.xlim(1000,6000)
plt.ylim(10,60)
plt.xlabel('Area of House')
plt.ylabel('Price of House')
plt.title('Area Vs Price of the House')
plt.rcParams['figure.dpi']=300  
```

- Xticks: Rotate x labels
- Yticks: Rotate y labels

```python
area=[1000,1520,3000,4500,5200]
price=[20,25,35,42,55]
plt.plot(area,price,'darkred',linewidth=2.5)
plt.xticks(rotation=45)
plt.yticks(rotation=10)
plt.xlabel('Area of House')
plt.ylabel('Price of House')
plt.title('Area Vs Price of the House')
plt.rcParams['figure.dpi']=300 
```

```python
x=np.arange(1,20)
y=np.sqrt(x)
z=np.log(x)
```

- Multiple plots: Multiple line in one chart called multiple plot
- Legend: Gives label to the chart
- Color: Used to give colors to chart

```python
plt.plot(x,y,'darkblue',label='Square Root',linewidth=2, linestyle='--')
plt.plot(x,z,'darkred',label='Log',linewidth=2)
plt.legend()
plt.title('Square Root Vs Log')
```

- it is a graph showing the number of observation within each given interval.

```python
plt.hist(df['width'])
plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['figure.dpi']=300
plt.title('Width Frequency histogram')
plt.xlabel('Width')
plt.ylabel('Frequency')
```

```python
df['width'].hist(bins=15)
plt.axvline(df['width'].mean(),color='r')
plt.axvline(df['width'].median(),color='y')
plt.xlabel('Frequency')
plt.ylabel('Width')
```
